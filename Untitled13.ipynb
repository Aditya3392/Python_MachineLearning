```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Load and preprocess the image
input_image_path = '/content/image_9989.png'  # Replace with the actual path
img = image.load_img(input_image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Normalize pixel values (if not already normalized)

# Convert the image to grayscale
gray = cv2.cvtColor(img_array[0], cv2.COLOR_BGR2GRAY)

# Normalize pixel values to be in the range [0, 1]
gray = cv2.normalize(gray, None, 0, 1, cv2.NORM_MINMAX)

# Edge detection using Canny
edges = cv2.Canny((gray * 255).astype(np.uint8), 50, 150)

# Display the processed images
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')

plt.show()
```
